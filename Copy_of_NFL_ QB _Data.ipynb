```python
# Load content from the URL for QB stats vs zone coverage
r_vs_zone = requests.get("https://sumersports.com/players/quarterback/?refinement=vs_zone")
soup_vs_zone = BeautifulSoup(r_vs_zone.text, 'html.parser') # Convert to a BS object

# Find the table containing QB data for vs_zone
qb_table_vs_zone = soup_vs_zone.find('table')

# Extract table headers for vs_zone
headers_vs_zone = [header.text.strip() for header in qb_table_vs_zone.find_all('th')]
print("Headers:", headers_vs_zone)

# Extract table rows for vs_zone
rows_vs_zone = []
for row in qb_table_vs_zone.find_all('tr')[1:]: # Skip header row
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    rows_vs_zone.append(row_data)

print("\nFirst 5 rows of data:", rows_vs_zone[:5])
```

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Load QB content
r = requests.get("https://sumersports.com/players/quarterback/")
soup = BeautifulSoup(r.text, 'html.parser') # Convert to a BS object
# print out HTML

print(soup.prettify())
```

```python
# Find the table containing QB data
qb_table = soup.find('table')

# Extract table headers
headers = [header.text.strip() for header in qb_table.find_all('th')]
print("Headers:", headers)

# Extract table rows
rows = []
for row in qb_table.find_all('tr')[1:]: # Skip header row
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    rows.append(row_data)

print("\nFirst 5 rows of data:", rows[:5])
```

```python
# Create a pandas DataFrame from the scraped data
df = pd.DataFrame(rows, columns=headers)

# Remove the leading number and period from the 'Player Name' column
df['Player Name'] = df['Player Name'].str.replace(r'^\d+\.\s*', '', regex=True)

# Drop the 'Season', 'Plays', and 'Team' columns
df = df.drop(columns=['Season', 'Plays', 'Team'])

# Display the entire cleaned DataFrame
display(df)
```
